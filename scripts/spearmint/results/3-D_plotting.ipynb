```python
def plotfig(iteridx):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    plt.xlim([16,50])
    plt.ylim([10,60])
    ax.set_zlim([2000,25000])
    Axes3D.scatter(ax, xs=itdf["k"][iteridx], ys=itdf["l"][iteridx], zs=itdf["N50"][iteridx])
```
