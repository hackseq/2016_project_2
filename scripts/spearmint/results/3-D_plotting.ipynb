```python
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.gca(projection='3d')
#ax = fig.add_subplot(111, projection='3d')
X, Y = np.meshgrid(itdf["k"], itdf["l"])
surf=ax.plot_surface(X=X, Y=Y, Z=itdf["N50"],rstride=1, cstride=1,  cmap=cm.coolwarm)
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

```python
def plotfig(iteridx):
    fig = plt.figure()
    #ax = fig.add_subplot(111, projection='3d')
    ax = fig.gca(projection='3d')
    plt.xlim([16,50])
    plt.ylim([10,60])
    plt.xlabel("k")
    plt.ylabel("l")
    ax.set_zlim([2000,25000])
    #ax = fig.add_subplot(111, projection='3d')
    X, Y = np.meshgrid(itdf["k"], itdf["l"])
    surf=ax.plot_surface(X=X, Y=Y, Z=itdf["N50"], rstride=3, cstride=3, alpha=0.4, cmap=cm.coolwarm,
                       linewidth=0, antialiased=True)
    fig.colorbar(surf, shrink=0.5, aspect=5)
    ax.scatter(xs=itdf["k"][iteridx], ys=itdf["l"][iteridx], zs=itdf["N50"][iteridx])
    plt.show()
```

```python
from ipywidgets import widgets, interact

interact(plotfig, iteridx=(1,39))
```
